```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(df_features, df_target, random_state=33, test_size=0.3)
```

```python
print(len(X_test), len(Y_test))
```

```python
model = RandomForestClassifier()
```

```python
model.fit(X_train, Y_train)
```

```python
y_pred = model.predict(X_test)
print(len(y_pred))
```

```python
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
score = classification_report( Y_test, y_pred)
```

```python
print(score)
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
params = {
    'criterion':['gini','entropy'],
    'max_depth':[1,2,3,4,5,6],
    'min_samples_split':[3,4,5,6],
    'min_samples_leaf':[1,2,3,4,5,6],
    'max_features':['auto', "sqrt", "log2"]
#     'bootstrap':True
}
```

```python
rv = RandomizedSearchCV(estimator=model, param_distributions=params,n_iter=10,cv=5)
```

```python
rv.fit(X_train, Y_train)
```

```python
rv.best_params_
```

```python
y_pred = rv.predict(X_test)
```

```python
score = accuracy_score(Y_test, y_pred)
```

```python
print(score)
```

```python
clss = classification_report(Y_test, y_pred)
```

```python
print(clss)
```

```python
cf = confusion_matrix(Y_test, y_pred)
```

```python
print(cf)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.metrics import RocCurveDisplay
```

```python
roc_auc_score(Y_test, y_pred)
```

```python
from sklearn import metrics

fpr, tpr, _ = roc_curve(Y_test, y_pred, pos_label=rv.classes_[1])
roc_display = RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
```
